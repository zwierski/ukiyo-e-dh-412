```python
dic_={}

for description in data['Description']:    
    nouns=[]    
    try: 
        description=str(description).lower()
        english_pattern = re.compile(r'[a-zA-Z]+')
        description=" ".join(english_pattern.findall(str(description)))
        is_noun = lambda pos: pos[:2] == 'NN'
        tokenized = nltk.word_tokenize(description)
        nouns=[word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 
    
        for n in nouns:
            if n not in dic_.keys():        
                dic_[n]=1
            else:
                dic_[n]=dic_[n]+1
    except TypeError:
        continue
```

```python
ordered_dic=sorted(dic_.items(),key = lambda x:x[1],reverse = True)        
print(ordered_dic)
```
